```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
import torch.utils.data as Data
class MLPregression(nn.Module):
    def __init__(self):
        super(MLPregression, self).__init__()
        self.hidden1 = nn.Linear(in_features=300, out_features=150, bias=True)
        self.drop1 = nn.Dropout(0.4)
        self.predict = nn.Linear(150, 1)

    def forward(self, x):
        x = F.relu(self.drop1(self.hidden1(x)))
        output = self.predict(x)
        return output[:, 0]
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
X_train_s = scale.fit_transform(X_train)
X_test_s = scale.transform(X_test)

X_train_t = torch.from_numpy(X_train_s.astype(np.float32))
y_train_t = torch.from_numpy(y_train.astype(np.float32))
X_test_t = torch.from_numpy(X_test_s.astype(np.float32))
y_test_t = torch.from_numpy(y_test.astype(np.float32))
# 转换为Dataloader
train_data = Data.TensorDataset(X_train_t, y_train_t)
test_data = Data.TensorDataset(X_test_t, y_test_t)
train_loader = Data.DataLoader(dataset = train_data, batch_size = 64, 
                               shuffle = True, num_workers = 1)
```

```python
import math
testnet = MLPregression()
optimizer = torch.optim.SGD(testnet.parameters(), lr = 0.01)
loss_func = nn.MSELoss() # 
train_loss_all = []
for epoch in range(25):
    train_loss = 0
    train_num = 0
    for step, (b_x, b_y) in enumerate(train_loader):
        output = testnet(b_x) 
        loss = loss_func(output, b_y)
        #loss=torch.tensor(math.sqrt(loss))
        optimizer.zero_grad() 
        loss.backward() 
        optimizer.step() 
        train_loss += loss.item() * b_x.size(0)
        train_num += b_x.size(0)
    print('Epoch{}   Train loss is_{}'.format(epoch,train_loss / train_num))
    train_loss_all.append(train_loss / train_num)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (8, 6))
plt.plot(train_loss_all, 'ro-', label = 'Train loss')
plt.legend()
plt.grid()
plt.xlabel('epoch')
plt.ylabel('Loss')
plt.show()
```
